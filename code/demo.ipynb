```python
import numpy as np
import matplotlib.pyplot as plt

# 定义直线拟合函数
def linear_regression(x, y): 
    N = len(x)
    sumx = x.sum()
    sumy = y.sum()
    sumx2 = (x**2).sum()
    sumxy = (x*y).sum()

    A = np.mat([[N, sumx], [sumx, sumx2]])
    b = np.array([sumy, sumxy])
 
    return np.linalg.solve(A, b)

X,Y = np.array(xmn),np.array(sum_lst)
a10, a11 = linear_regression(X, Y)
 
# 生成拟合直线的绘制点
_Y1 = np.array([a10 + a11 * x for x in X])

# 画图
plt.figure(figsize=(6,6))
plt.scatter(X,Y)
plt.plot(X, _Y1, 'red', linewidth=2) 
#plt.title("y = {} + {}x".format(round(a10,4), round(a11,4))) # 参数保留4位小数
plt.title("y = {} + {}x".format(a10, a11)) # 标题
plt.show()
```
